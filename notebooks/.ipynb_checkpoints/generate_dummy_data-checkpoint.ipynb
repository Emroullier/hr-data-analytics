# Generate dummy input - predict leaving

```python
# Generate a dataset of 'number' random rows from original dataset
number = 10

def generate_dummy_data_leaving(number, data):
    '''
    generate_data_leaving creates a json file of 'number' rows from data and saves it locally.
    For demonstration purpose with Front-End.
    
    '''
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    file_path = os.path.join('../raw_data/dummy_data/',f"{timestamp}_dummy_{number}.json")
    data.sample(number).to_json(file_path)

# generate_dummy_data_leaving(number,dataset)
```

# Generate dummy input - predict hiring

```python
# Generate a dataset of 'number' random rows *** for a specific position ***
# from original dataset
number = 3

def generate_dummy_data_hiring(number, data):
    '''
    generate_data_leaving creates a json file of 'number' random rows *** for a specific position***
    from data and saves it locally.
    For demonstration purpose with Front-End.
    
    '''
    shape=0
    
    while shape < number:    
        # Random selection for Department, GEO, average_montly_hours, salary and Role
        rand_Department = np.random.choice(dataset['Department'].unique(),1)
        rand_GEO = np.random.choice(dataset['GEO'].unique(),1)
        rand_average_montly_hours = np.random.choice(dataset['average_montly_hours'].unique(),1)
        rand_salary = np.random.choice(dataset['salary'].unique(),1)
        rand_Role = np.random.choice(dataset['Role'].unique(),1)

        query = f"@dataset['Department']=='{rand_Department[0]}'and\
        @dataset['GEO']== '{rand_GEO[0]}'and\
        @dataset['average_montly_hours'] > {rand_average_montly_hours[0]} and\
        @dataset['salary']=='{rand_salary[0]}'and\
        @dataset['Role']=='{rand_Role[0]}'"
        input_features = dataset.query(query)
        
        if input_features.shape[0] >= number:   
            #Create and save json file locally
            timestamp = time.strftime("%Y%m%d_%H%M%S")
            file_path = os.path.join('../raw_data/dummy_data/',f"{timestamp}_dummy_{rand_Role[0]}_{rand_Department[0]}_{rand_GEO[0]}_{rand_average_montly_hours[0]}_{rand_salary[0]}_{number}.json")
            input_features.sample(number).to_json(file_path)
            break
        else : 
            shape = input_features.shape[0]

# generate_dummy_data_hiring(number,dataset)
```
