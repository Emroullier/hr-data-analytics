## Time spent at company

```python
feature_variable = "time_spend_company"
n_bins=8

ax = (data
      .loc[lambda x: x["left_Company"] == 0]
      [feature_variable]
      .plot
      .hist(label="Did Not Leave Company",density=True,bins=n_bins,alpha=0.3)
     )

(data
 .loc[lambda x: x["left_Company"] == 1]
 ["time_spend_company"]
 .plot
 .hist(ax=ax, label="Left Company",density=True,bins=n_bins,alpha=0.3)
)

plt.xlabel(feature_variable)
plt.legend()
```

```python
avg_time_spent = data.groupby('left_Company')['time_spend_company'].mean().reset_index()

avg_time_spent.columns = ['Left Company', 'Average Time Spent (years)']

plt.figure(figsize=(10, 6))
plt.bar(avg_time_spent['Left Company'], avg_time_spent['Average Time Spent (years)'], color=['green', 'red'])
plt.title('Average Time Spent at Company by Employee Status')
plt.xlabel('Employee Status (0 = Stayed, 1 = Left)')
plt.ylabel('Average Time Spent (years)')
plt.xticks([0, 1], ['Stayed', 'Left'])
plt.grid(axis='y')

for index, value in avg_time_spent.iterrows():
    plt.text(value['Left Company'], value['Average Time Spent (years)'] + 0.1, f'{value["Average Time Spent (years)"]:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

It seems that the longer the people stay the more likely they are to leave. Most likely to leave between 3 - 6 years.

## Gender, promotion, work accident

```python
data['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))  # 2 rows, 2 columns

labels = ['No', 'Yes']

gender_ct = pd.crosstab(data['Gender'], data['left_Company'])
gender_ct.plot(kind='bar', ax=axs[0, 0])
axs[0, 0].set_title('Gender Distribution')
axs[0, 0].set_xlabel('Gender')
axs[0, 0].set_ylabel('Count')
axs[0, 0].legend(title='Left Company', labels=labels)

promotion_ct = pd.crosstab(data['promotion_last_5years'], data['left_Company'])
promotion_ct.plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Promotions in Last 5 Years')
axs[0, 1].set_xlabel('Promotion in Last 5 Years')
axs[0, 1].set_ylabel('Count')
axs[0, 1].legend(title='Left Company', labels=labels)

accident_ct = pd.crosstab(data['Work_accident'], data['left_Company'])
accident_ct.plot(kind='bar', ax=axs[1, 0])
axs[1, 0].set_title('Work Accident')
axs[1, 0].set_xlabel('Work Accident')
axs[1, 0].set_ylabel('Count')
axs[1, 0].legend(title='Left Company', labels=labels)

rising_star_ct = pd.crosstab(data['Rising_Star'], data['left_Company'])
rising_star_ct.plot(kind='bar', ax=axs[1, 1])
axs[1, 1].set_title('Rising Star Status')
axs[1, 1].set_xlabel('Rising Star')
axs[1, 1].set_ylabel('Count')
axs[1, 1].legend(title='Left Company', labels=labels)

plt.tight_layout()
plt.show()
```

Gender has no influence on people leaving the company. If promoted or rising star, seems much less likely to leave. 

## Attrition by department

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Department', data=data)
plt.title('Distribution of Departments')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
dept_ct = pd.crosstab(data['Department'], data['left_Company'])

dept_ct['Total'] = dept_ct.sum(axis=1)
dept_ct['Percentage Left'] = (dept_ct[1] / dept_ct['Total']) * 100
dept_ct['Percentage Stayed'] = (dept_ct[0] / dept_ct['Total']) * 100

dept_ct_sorted = dept_ct.sort_values(by=1, ascending=False)

fig, ax = plt.subplots(figsize=(12, 8))
bars = dept_ct_sorted[[0, 1]].plot(kind='bar', color=['green', 'red'], ax=ax, width=0.8)

for i, (perc_left, perc_stayed) in enumerate(zip(dept_ct_sorted['Percentage Left'], dept_ct_sorted['Percentage Stayed'])):
    ax.text(i - 0.2, dept_ct_sorted.iloc[i, 0] + 5, f'{perc_stayed:.1f}%', ha='center', va='bottom', color='black', fontsize=9)
    ax.text(i + 0.2, dept_ct_sorted.iloc[i, 1] + 5, f'{perc_left:.1f}%', ha='center', va='bottom', color='black', fontsize=9)

plt.title('Attrition by department')
ax.set_xlabel('Department')
ax.set_ylabel('Number of Employees')
ax.set_xticklabels(dept_ct_sorted.index, rotation=45)
plt.legend(title='Left Company', labels=['Stayed', 'Left'])

plt.tight_layout()
plt.show()
```

No influence of Department on attrition.

## Attrition by geolocalisation

```python
geo_ct = pd.crosstab(data['GEO'], data['left_Company'])

geo_ct['Total'] = geo_ct.sum(axis=1)
geo_ct['Percentage Left'] = (geo_ct[1] / geo_ct['Total']) * 100
geo_ct['Percentage Stayed'] = (geo_ct[0] / geo_ct['Total']) * 100

geo_ct_sorted = geo_ct.sort_values(by=1, ascending=False)

fig, ax = plt.subplots(figsize=(12, 8))
bars = geo_ct_sorted[[0, 1]].plot(kind='bar', color=['green', 'red'], ax=ax, width=0.8)

for i, (perc_left, perc_stayed) in enumerate(zip(geo_ct_sorted['Percentage Left'], geo_ct_sorted['Percentage Stayed'])):
    ax.text(i - 0.2, geo_ct_sorted.iloc[i, 0] + 5, f'{perc_stayed:.1f}%', ha='center', va='bottom', color='black', fontsize=9)
    ax.text(i + 0.2, geo_ct_sorted.iloc[i, 1] + 5, f'{perc_left:.1f}%', ha='center', va='bottom', color='black', fontsize=9)

plt.title('Attrition by Geographic Region')
ax.set_xlabel('Geographic Region')
ax.set_ylabel('Number of Employees')
ax.set_xticklabels(geo_ct_sorted.index, rotation=45)
plt.legend(title='Left Company', labels=['Stayed', 'Left'])

plt.tight_layout()
plt.show()
```

Not much difference between geographical locations in terms of attrition.

## Employee Satisfaction - on premise

```python
data['left_Company'] = data['left_Company'].astype(str)

satisfaction_columns = ['EMP_Sat_OnPrem_1','EMP_Sat_OnPrem_2', 'EMP_Sat_OnPrem_3', 'EMP_Sat_OnPrem_4', 'EMP_Sat_OnPrem_5']
avg_satisfaction = data.groupby('left_Company')[satisfaction_columns].mean().transpose()

plt.figure(figsize=(12, 6))

for status in avg_satisfaction.columns:
    plt.plot(avg_satisfaction.index, avg_satisfaction[status], marker='o', label='Left' if status == '1' else 'Stayed')

plt.title('Average On-Premise Employee Satisfaction by Employee Status')
plt.xlabel('Satisfaction Measure')
plt.ylabel('Average Satisfaction')
plt.legend(title='Employee Status')
plt.grid(True)
plt.show()
```

```python
data['left_Company'] = data['left_Company'].astype(str)

remote_satisfaction_columns = ['EMP_Sat_Remote_1', 'EMP_Sat_Remote_2', 'EMP_Sat_Remote_3', 'EMP_Sat_Remote_4', 'EMP_Sat_Remote_5']
avg_remote_satisfaction = data.groupby('left_Company')[remote_satisfaction_columns].mean().transpose()

plt.figure(figsize=(12, 6))

for status in avg_remote_satisfaction.columns:
    plt.plot(avg_remote_satisfaction.index, avg_remote_satisfaction[status], marker='o', label='Left' if status == '1' else 'Stayed')

plt.title('Average Remote Employee Satisfaction by Employee Status')
plt.xlabel('Satisfaction Measure over time')
plt.ylabel('Average Satisfaction')
plt.legend(title='Employee Status')
plt.grid(True)
plt.show()
```

Those who leave tend to be least satisfied with remote working and more satisfied with working on premise. 

```python
data['left_Company'] = data['left_Company'].astype(str)

# Calculate the average for each on-premise satisfaction column, grouped by 'left_Company'
onprem_satisfaction_columns = ['EMP_Sat_OnPrem_1','EMP_Sat_OnPrem_2', 'EMP_Sat_OnPrem_3', 'EMP_Sat_OnPrem_4', 'EMP_Sat_OnPrem_5']
avg_onprem_satisfaction = data.groupby('left_Company')[onprem_satisfaction_columns].mean().transpose()

# Calculate the average for each remote satisfaction column, grouped by 'left_Company'
remote_satisfaction_columns = ['EMP_Sat_Remote_1', 'EMP_Sat_Remote_2', 'EMP_Sat_Remote_3', 'EMP_Sat_Remote_4', 'EMP_Sat_Remote_5']
avg_remote_satisfaction = data.groupby('left_Company')[remote_satisfaction_columns].mean().transpose()

fig, axs = plt.subplots(1, 2, figsize=(20, 6), sharey=True)

# On-Premise Satisfaction
for status in avg_onprem_satisfaction.columns:
    axs[0].plot(avg_onprem_satisfaction.index, avg_onprem_satisfaction[status], marker='o', label='Left' if status == '1' else 'Stayed')
axs[0].set_title('Average On-Premise Employee Satisfaction by Employee Status')
axs[0].set_xlabel('On-Premise Satisfaction Measure over time')
axs[0].set_ylabel('Average Satisfaction')
axs[0].legend(title='Employee Status')
axs[0].grid(True)

# Remote Satisfaction
for status in avg_remote_satisfaction.columns:
    axs[1].plot(avg_remote_satisfaction.index, avg_remote_satisfaction[status], marker='o', label='Left' if status == '1' else 'Stayed')
axs[1].set_title('Average Remote Employee Satisfaction by Employee Status')
axs[1].set_xlabel('Remote Satisfaction Measure over time')
axs[1].legend(title='Employee Status')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
