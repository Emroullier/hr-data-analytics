# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
```

# Retrieve data

```python
dataset = pd.read_csv('../raw_data/HR_Engagement_Sat_Sales_UpdatedV4.0.csv')
```

```python
pd.set_option('display.max_columns', None)
print(dataset.shape)
dataset.head()
```

# Drop Categorical columns

```python
#Data types of dataset
dataset.dtypes.value_counts()
```

```python
# Select only numerical columns in dataset
dataset = dataset.select_dtypes(exclude=[object])
print(dataset.shape)
dataset.head()
```

```python
#Data types of dataset
dataset.dtypes.value_counts()
```

# Simple Imputer

```python
# Check if the columns contain NA values, null values, missing values
pd.set_option('display.max_rows', None)
(dataset.count()/len(dataset)*100).sort_values(ascending=True)
```

```python
#Data types of dataset
dataset.dtypes.value_counts()
```

```python
#Saving dataset types in a variable (SimpleImputer modify dtypes when transforming)
dataset_dtypes = dataset.dtypes
```

```python
# Instanciate imputer
imputer = SimpleImputer(strategy="most_frequent").set_output(transform="pandas")
```

```python
# Call the "fit" method on the object
imputer.fit(dataset) 
```

```python
# Call the "transform" method on the object
dataset = imputer.transform(dataset)
dataset = dataset.astype(dataset_dtypes)
print(dataset.shape)
dataset.head()
```

```python
#Data types of dataset
dataset.dtypes.value_counts()
```

```python
# No NA values, null values, missing values thanks to SimpleImputer
(dataset.count()/len(dataset)*100).sort_values(ascending=True)
```


# Define X (features)


```python
#Features
X = dataset.drop(columns=['left_Company', 'ID'])
print(X.shape)
X.head()
```

```python
#Data types of X
X.dtypes.value_counts()
```

```python
# ********************************
```

```python
# Check for constant values <=> standard deviation == 0
zero_std_cols = X.std(axis=0, numeric_only=True)[X.std(axis=0, numeric_only=True) == 0]
zero_std_cols;
```

```python
#Check the mean value of the columns with standard deviation == 0
unique_value_cols = X.mean(axis=0, numeric_only=True)[zero_std_cols.index]
unique_value_cols;
```

```python
unique_value_cols_names = unique_value_cols.index.to_list()
unique_value_cols_names
```

```python
X = X.drop(columns=unique_value_cols_names)
print(X.shape)
X.head()
```

```python
# ********************************
```

# Define y (target)

```python
#Target
y = dataset['left_Company']
print(y.shape)
y.head()
```

```python
y.value_counts()/len(y)*100
```

# Quick Logistic regression

```python
#Instanciating SMOTE
smote = SMOTE()
smote
```

```python
#Oversampling features and target
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Check that target resampled train set is imbalanced
y_resampled.value_counts()/len(y_resampled)*100
```

```python
print("\naccuracy all initial features")
print(cross_val_score(LogisticRegression(), X_resampled, y_resampled, cv=5).mean())
```

```python
mdl = LogisticRegression()
mdl.fit(X,y)
cv_results = cross_validate(mdl,
                            X, 
                            y, 
                            cv=5, 
                            scoring=['precision'])

score = cv_results['test_precision'].mean()
score
```

# Standardizing X

⚠️ Data must be centered around its mean before applying PCA ⚠️

```python
# Instanciate a Standard Scaler
std_scaler = StandardScaler().set_output(transform="pandas")
std_scaler
```

```python
# Fittng Standard Scaler to X
std_scaler.fit(X)
```

```python
X_std = std_scaler.transform(X) 
print(X_std.shape)
X_std.head()
```

```python
#Data types of X_std
X_std.dtypes.value_counts()
```

# Computing Principal Components

```python
# Instanciate PCA
pca = PCA()
```

```python
#Fitting PCA to scaled features
pca.fit(X_std)
```

```python
# Access our PCs 
W = pca.components_

# Print PCs as COLUMNS
W = pd.DataFrame(W.T,
                 index=X_std.columns,
                 columns=[f'PC{i}' for i in range(1, len(W)+1)])
W
```

# Project our dataset into this new space of PCs

```python
#Project our dataset into this new space of PCs
X_proj = pca.transform(X_std)
X_proj = pd.DataFrame(X_proj, columns=[f'PC{i}' for i in range(1, len(W)+1)])
print(X_proj.shape)
X_proj.head()
```

# Explained variance

```python
#Explained variance
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Principal Component'); plt.ylabel('% explained variance');
plt.grid(visible=True)
```

# PCA for Dimensionality Reduction

```python
# PCA for dimensionality reduction
plt.figure(figsize=(4,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.ylim(ymin=0)
plt.title('cumulated share of explained variance')
plt.xlabel('# of principal component used');
plt.grid(visible=True)
```

# Test Model Performance with k Dimensions)

```python
# Fit a PCA with specific number of components
n_components = 15
pca_perf = PCA(n_components=n_components).fit(X_std)
```

```python
# Project the data into specific number of  dimensions
X_proj_perf = pd.DataFrame(pca_perf.fit_transform(X_std), columns=[f'PC{i}' for i in range(1, n_components+1)])
print(X_proj_perf.shape)
X_proj_perf.head()
```

```python
print(f"accuracy {n_components} PCs")
print(cross_val_score(LogisticRegression(), X_proj_perf, y, cv=5).mean())

print("\naccuracy all initial features")
print(cross_val_score(LogisticRegression(), X_std, y, cv=5).mean())
```
