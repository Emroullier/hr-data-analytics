# Generate dummy input - predict leaving

```python
# Drop target from dataset
dataset.drop(columns=['left_Company'], inplace=True)
```

```python
def generate_dummy_data(number, data):
    '''
    generate_data_leaving creates a json file of 'number' rows from data and saves it locally.
    For demonstration purpose with Front-End.
    
    '''
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    file_path = os.path.join('../raw_data/dummy_data/',f"{timestamp}_dummy_{number}.json")
    data.sample(number).to_json(file_path)
    
generate_dummy_data(10,dataset)
```
